```python
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split

poke_data = pd.read_csv("pokemon.csv")
poke_feature = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']

X = poke_data[poke_feature]
y = poke_data.Legendary

# Split train data and test data
trainX, testX, trainY, testY = train_test_split(X,y,random_state = 0)

print(trainX)
print(trainY)
```

```python
# define model
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(6,)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(128,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64,activation='relu',kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
# complie model
model.compile(optimizer='adam',loss=tf.keras.losses.BinaryCrossentropy(),metrics=['accuracy'])

# train model
model.fit(trainX,trainY, epochs=200, batch_size=200,shuffle=True, validation_split=0.2)

#evaluate
print(model.evaluate(trainX,trainY))
print(model.evaluate(testX,testY))
```

```python
# predict
print(tf.round(model.predict(trainX[200:225])))
print(trainY[200:225])
```
